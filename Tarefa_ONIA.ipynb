```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/1 b informatica/Cópia de heart_attack_prediction_dataset.csv")

df.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure()
plt.hist(df['Age'], bins=20 )
plt.xlabel('idade')
plt.ylabel('quantidade')
plt.title('distribuição de idade')
plt.show()
```

```python
media_colesterol = df.groupby('Age')['Cholesterol'].mean()

plt.figure()
plt.bar(media_colesterol.index, media_colesterol.values)
plt.xlabel('idade')
plt.ylabel('colesterol')
plt.title('idade vs colesterol')
plt.show()
```

```python
contagem = df['Heart Attack Risk'].value_counts()

plt.figure()
plt.pie(contagem, labels=contagem.index, autopct='%1.1f%%')
plt.title('Proporção de Risco de Ataque Cardíaco')
plt.show()
```

```python
media_imc = df.groupby('Sex')['BMI'].mean()

plt.figure()
plt.bar(media_imc.index, media_imc.values)

plt.xlabel('Sexo')
plt.ylabel('IMC Médio')
plt.title('IMC Médio por Sexo')

plt.show()
```

```python
idade_media = df.groupby('Sex')['Age'].mean()

plt.figure()
plt.plot(idade_media.index, idade_media.values)

plt.xlabel('Sexo')
plt.ylabel('Idade Média')
plt.title('Idade Média por Sexo')

plt.show()
```
